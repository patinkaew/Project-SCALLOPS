# Comparison of petawatt amplification (Glass front end and OPA front end)

```python
import matplotlib.pyplot as plt
```

```python
# read stored data
%store -r cross_section

%store -r initial_spectrum_glass_plot
%store -r initial_pulse_glass_plot
%store -r final_spectrum_glass_plot
%store -r final_spectrum_normalized_glass_plot
%store -r final_pulse_glass_plot
%store -r final_normalized_pulse_glass_plot

%store -r initial_spectrum_opa_plot
%store -r initial_pulse_opa_plot
%store -r final_spectrum_opa_plot
%store -r final_spectrum_normalized_opa_plot
%store -r final_pulse_opa_plot
%store -r final_normalized_pulse_opa_plot
```

```python
plt.plot(*initial_spectrum_glass_plot, label = 'Glass')
plt.plot(*initial_spectrum_opa_plot, label = 'OPA')
plt.xlim((1035, 1075))
plt.legend()
plt.title('Input spectrum', fontsize = 14)
plt.xlabel('wavelength (nm)', fontsize = 12)
plt.ylabel('power (W)', fontsize = 12)
plt.grid(alpha = 0.5)
# plt.savefig('../pics/pw_input_spectrum.png', quality = 95, dpi = 600, bbox_inches='tight')
plt.show()
```

```python
fig, ax1 = plt.subplots()

lns1 = ax1.plot(*initial_spectrum_glass_plot, label = 'Glass')
lns2 = ax1.plot(*initial_spectrum_opa_plot, label = 'OPA')
ax1.set_xlabel('wavelength (nm)', fontsize = 12)
ax1.set_ylabel('power (W)', fontsize = 12)
ax1.set_xlim((1025, 1100))
ax1.grid(alpha = 0.5)
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

lns3 = ax2.plot(*cross_section, 'g--', label = 'cross section')
ax2.set_ylabel('cross section (1/m^2)', fontsize = 12)

leg = lns1 + lns2 + lns3
labs = [l.get_label() for l in leg]
ax1.legend(leg, labs, loc=0)

fig.suptitle('Input spectrum and amplifer cross section', fontsize = 14)
plt.show()
```

```python
plt.plot(*initial_pulse_glass_plot, label = 'Glass')
plt.plot(*initial_pulse_opa_plot, label = 'OPA')
plt.xlim((-400e-15, 400e-15))
plt.legend()
plt.title('Input pulse', fontsize = 14)
plt.xlabel('time (s)', fontsize = 12)
plt.ylabel('power (W)', fontsize = 12)
plt.grid(alpha = 0.5)
# plt.savefig('../pics/pw_input_pulse.png', quality = 95, dpi = 600, bbox_inches='tight')
plt.show()
```

```python
plt.plot(*final_spectrum_glass_plot, label = 'Glass')
plt.plot(*final_spectrum_opa_plot, label = 'OPA')
plt.legend()
plt.xlim((1040, 1080))
plt.title('Output spectrum', fontsize = 14)
plt.xlabel('wavelength (nm)', fontsize = 12)
plt.ylabel('power (W)', fontsize = 12)
plt.grid(alpha = 0.5)
# plt.savefig('../pics/pw_output_spectrum.png', quality = 95, dpi = 600)
plt.show()
```
